```python
""" ?뺤씤 """
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import xarray as xr

day_path = r"/mnt/d/project/UmaxFormer/data/raw/era5/uas.day.195801-202512.nc"
mon_path = r"/mnt/d/project/UmaxFormer/data/raw/era5/uas.mon.195801-202512.nc"
anom_mon_path = r"/mnt/d/project/UmaxFormer/data/interim/era5/uas.mon.195801-202512.anom_1x2.nc"
anom_monmax_path = r"/mnt/d/project/UmaxFormer/data/interim/era5/uasmax.mon.195801-202512.anom_1x2.nc"

def load_hovmoller(path, t0, t1, lon1=140, lon2=180, lat1=-5, lat2=5):
    ds = xr.open_dataset(path)

    rename_dict = {}

    if "valid_time" in ds.coords:
        rename_dict["valid_time"] = "time"
    elif "Time" in ds.coords:
        rename_dict["Time"] = "time"

    if "lon" in ds.coords:
        rename_dict["lon"] = "longitude"
    elif "LONGITUDE" in ds.coords:
        rename_dict["LONGITUDE"] = "longitude"

    if "lat" in ds.coords:
        rename_dict["lat"] = "latitude"
    elif "LATITUDE" in ds.coords:
        rename_dict["LATITUDE"] = "latitude"

    if rename_dict:
        ds = ds.rename(rename_dict)

    for coord in ["time", "longitude", "latitude"]:
        if coord not in ds.coords:
            raise KeyError(
                f"{path} ?먯꽌 '{coord}' 醫뚰몴瑜?李얠쓣 ???놁뒿?덈떎. "
                f"?꾩옱 coords = {list(ds.coords)}"
            )

    ds = ds.sortby("latitude")
    ds = ds.sel(
        time=slice(t0, t1),
        longitude=slice(lon1, lon2),
        latitude=slice(lat1, lat2),
    )

    # 蹂???먮룞 ?좏깮
    if "uas" in ds.data_vars:
        da = ds["uas"]
    elif "uasmax" in ds.data_vars:
        da = ds["uasmax"]
    else:
        varname = list(ds.data_vars)[0]
        da = ds[varname]

    if da.sizes["latitude"] == 0:
        raise ValueError(f"{path}: latitude ?좏깮 寃곌낵媛 鍮꾩뿀?듬땲??")
    if da.sizes["longitude"] == 0:
        raise ValueError(f"{path}: longitude ?좏깮 寃곌낵媛 鍮꾩뿀?듬땲??")

    w = np.cos(np.deg2rad(da["latitude"]))
    hov = da.weighted(w).mean("latitude")
    return hov

def plot_year(year):
    hov_day_raw = load_hovmoller(day_path, f"{year}-01-01", f"{year}-12-31")
    hov_mon_native = load_hovmoller(mon_path, f"{year}-01-01", f"{year}-12-31")
    hov_mon_anom = load_hovmoller(anom_mon_path, f"{year}-01-01", f"{year}-12-31")
    hov_monmax_anom = load_hovmoller(anom_monmax_path, f"{year}-01-01", f"{year}-12-31")

    hov_day_monthly_mean = hov_day_raw.groupby("time.month").mean("time")
    hov_day_monthly_max  = hov_day_raw.groupby("time.month").max("time")
    hov_mon_native_by_month = hov_mon_native.groupby("time.month").mean("time")
    hov_mon_anom_by_month = hov_mon_anom.groupby("time.month").mean("time")
    hov_monmax_anom_by_month = hov_monmax_anom.groupby("time.month").mean("time")

    # ?곕룄蹂??꾩껜 怨듯넻 color scale
    vabs_all = max(
        float(np.nanmax(np.abs(hov_day_raw.values))),
        float(np.nanmax(np.abs(hov_day_monthly_max.values))),
        float(np.nanmax(np.abs(hov_mon_native_by_month.values))),
        float(np.nanmax(np.abs(hov_mon_anom_by_month.values))),
        float(np.nanmax(np.abs(hov_monmax_anom_by_month.values))),
    )
    levels_all = np.linspace(-vabs_all, vabs_all, 21)

    month_labels = ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]
    xticks = [140, 150, 160, 170, 180]
    xticklabels = ["140E", "150E", "160E", "170E", "180"]

    fig, axes = plt.subplots(1, 5, figsize=(30, 8), constrained_layout=True)

    # (1) daily raw
    cf1 = axes[0].contourf(
        hov_day_raw["longitude"],
        hov_day_raw["time"],
        hov_day_raw,
        levels=levels_all,
        cmap="RdBu_r",
        extend="both"
    )
    axes[0].set_title(f"{year} Daily Raw uas\n5째S??째N average, 140째E??80째")
    axes[0].set_xlabel("Longitude")
    axes[0].set_ylabel("Time")
    axes[0].set_xticks(xticks)
    axes[0].set_xticklabels(xticklabels)
    axes[0].yaxis.set_major_locator(mdates.MonthLocator(interval=1))
    axes[0].yaxis.set_major_formatter(mdates.DateFormatter("%b"))
    fig.colorbar(cf1, ax=axes[0], pad=0.02).set_label("m/s")

    # (2) daily -> monthly max
    cf2 = axes[1].contourf(
        hov_day_monthly_max["longitude"],
        hov_day_monthly_max["month"],
        hov_day_monthly_max,
        levels=levels_all,
        cmap="RdBu_r",
        extend="both"
    )
    axes[1].set_title(f"{year} Daily ??Monthly Max uas\n5째S??째N average, 140째E??80째")
    axes[1].set_xlabel("Longitude")
    axes[1].set_ylabel("Month")
    axes[1].set_xticks(xticks)
    axes[1].set_xticklabels(xticklabels)
    axes[1].set_yticks(np.arange(1, 13))
    axes[1].set_yticklabels(month_labels)
    fig.colorbar(cf2, ax=axes[1], pad=0.02).set_label("m/s")

    # (3) native monthly mean
    cf3 = axes[2].contourf(
        hov_mon_native_by_month["longitude"],
        hov_mon_native_by_month["month"],
        hov_mon_native_by_month,
        levels=levels_all,
        cmap="RdBu_r",
        extend="both"
    )
    axes[2].set_title(f"{year} Native Monthly Mean uas\n5째S??째N average, 140째E??80째")
    axes[2].set_xlabel("Longitude")
    axes[2].set_ylabel("Month")
    axes[2].set_xticks(xticks)
    axes[2].set_xticklabels(xticklabels)
    axes[2].set_yticks(np.arange(1, 13))
    axes[2].set_yticklabels(month_labels)
    fig.colorbar(cf3, ax=axes[2], pad=0.02).set_label("m/s")

    # (4) monthly mean anomaly
    cf4 = axes[3].contourf(
        hov_mon_anom_by_month["longitude"],
        hov_mon_anom_by_month["month"],
        hov_mon_anom_by_month,
        levels=levels_all,
        cmap="RdBu_r",
        extend="both"
    )
    axes[3].set_title(f"{year} Monthly Mean uas Anomaly\n5째S??째N average, 140째E??80째")
    axes[3].set_xlabel("Longitude")
    axes[3].set_ylabel("Month")
    axes[3].set_xticks(xticks)
    axes[3].set_xticklabels(xticklabels)
    axes[3].set_yticks(np.arange(1, 13))
    axes[3].set_yticklabels(month_labels)
    fig.colorbar(cf4, ax=axes[3], pad=0.02).set_label("m/s")

    # (5) monthly max anomaly
    cf5 = axes[4].contourf(
        hov_monmax_anom_by_month["longitude"],
        hov_monmax_anom_by_month["month"],
        hov_monmax_anom_by_month,
        levels=levels_all,
        cmap="RdBu_r",
        extend="both"
    )
    axes[4].set_title(f"{year} Monthly Max uas Anomaly\n5째S??째N average, 140째E??80째")
    axes[4].set_xlabel("Longitude")
    axes[4].set_ylabel("Month")
    axes[4].set_xticks(xticks)
    axes[4].set_xticklabels(xticklabels)
    axes[4].set_yticks(np.arange(1, 13))
    axes[4].set_yticklabels(month_labels)
    fig.colorbar(cf5, ax=axes[4], pad=0.02).set_label("m/s")

    plt.show()


plot_year(1996)
plot_year(1997)
plot_year(2014)
plot_year(2015)
plot_year(2023)
plot_year(2024)
```
