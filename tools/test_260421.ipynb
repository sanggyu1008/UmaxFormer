```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================================================
# ?ㅼ젙
# =========================================================
uas_path = "/mnt/d/project/UmaxFormer/data/input/test_dataset_godas_era5_198001_202512.nc"

lat_range = (-5, 5)
lon_range = (140, 220)     # 140E ~ 140W

use_train_stats = False
train_stats_path = "/mnt/d/project/UmaxFormer/data/input/train_dataset.nc"

# =========================================================
# ?좏떥
# =========================================================
def prep_ds(ds):
    if "lon" in ds.coords and ds["lon"].min() < 0:
        ds = ds.assign_coords(lon=((ds["lon"] + 360) % 360)).sortby("lon")
    if "lat" in ds.coords and ds["lat"][0] > ds["lat"][-1]:
        ds = ds.sortby("lat")
    return ds

def ensure_time_name(obj):
    if "time" in obj.coords:
        return obj
    if "valid_time" in obj.coords:
        return obj.rename({"valid_time": "time"})
    if "month" in obj.coords:
        return obj.rename({"month": "time"})
    raise KeyError(f"No time coordinate found. coords={list(obj.coords)}")

def monthly_time_index(da):
    da = ensure_time_name(da)
    t = pd.to_datetime(da["time"].values)
    t = pd.to_datetime(pd.PeriodIndex(t, freq="M").to_timestamp())
    return da.assign_coords(time=t).sortby("time")

def flatten_valid(da):
    x = np.asarray(da.values).ravel()
    return x[np.isfinite(x)]

def compute_stats(da):
    vals = flatten_valid(da)
    return vals.mean(), vals.std()

def zscore(da, mean, std, eps=1e-12):
    return (da - mean) / max(std, eps)

def lat_weighted_lon_time(da):
    """
    da dims: time, lat, lon
    output dims: time, lon
    """
    w = np.cos(np.deg2rad(da["lat"]))
    return da.weighted(w).mean("lat")

def plot_hov(ax, da, title, vlim=None, cmap="RdBu_r"):
    # da dims: time, lon
    x = da["lon"].values
    y = pd.to_datetime(da["time"].values)
    if vlim is None:
        vlim = np.nanmax(np.abs(da.values))
    im = ax.pcolormesh(x, y, da.values, shading="auto", vmin=-vlim, vmax=vlim, cmap=cmap)
    ax.set_title(title)
    ax.set_xlabel("Longitude")
    ax.set_ylabel("Time")
    return im

# =========================================================
# 1) ?곗씠??濡쒕뱶
# =========================================================
ds = prep_ds(xr.open_dataset(uas_path))
uas = monthly_time_index(ds["uas"].sel(lat=slice(*lat_range), lon=slice(*lon_range)))
uasmax = monthly_time_index(ds["uasmax"].sel(lat=slice(*lat_range), lon=slice(*lon_range)))

print("=== source ===")
print("uas    :", uas.dims, uas.shape)
print("uasmax :", uasmax.dims, uasmax.shape)

# =========================================================
# 2) ?뺢퇋???듦퀎
# =========================================================
if use_train_stats:
    ds_stat = prep_ds(xr.open_dataset(train_stats_path))
    uas_stat = monthly_time_index(ds_stat["uas"].sel(lat=slice(*lat_range), lon=slice(*lon_range)))
    uasmax_stat = monthly_time_index(ds_stat["uasmax"].sel(lat=slice(*lat_range), lon=slice(*lon_range)))
    print(f"\nUsing TRAIN stats from {train_stats_path}")
else:
    uas_stat = uas
    uasmax_stat = uasmax
    print(f"\nUsing CURRENT stats from {uas_path}")

uas_mean, uas_std = compute_stats(uas_stat)
uasmax_mean, uasmax_std = compute_stats(uasmax_stat)

print(f"uas    mean={uas_mean:.6f}, std={uas_std:.6f}")
print(f"uasmax mean={uasmax_mean:.6f}, std={uasmax_std:.6f}")

# own normalization
uas_norm = zscore(uas, uas_mean, uas_std)
uasmax_norm = zscore(uasmax, uasmax_mean, uasmax_std)

# =========================================================
# 3) latitude-averaged Hovm철ller (time, lon)
# =========================================================
uas_hov = lat_weighted_lon_time(uas)
uasmax_hov = lat_weighted_lon_time(uasmax)

uas_norm_hov = lat_weighted_lon_time(uas_norm)
uasmax_norm_hov = lat_weighted_lon_time(uasmax_norm)

diff_hov = uasmax_hov - uas_hov
diff_norm_hov = uasmax_norm_hov - uas_norm_hov

# =========================================================
# 4) 泥?踰덉㎏ figure留?洹몃┝
# =========================================================
raw_vlim = np.nanmax(np.abs([
    np.nanmax(np.abs(uas_hov.values)),
    np.nanmax(np.abs(uasmax_hov.values)),
    np.nanmax(np.abs(diff_hov.values)),
]))

norm_vlim = np.nanmax(np.abs([
    np.nanmax(np.abs(uas_norm_hov.values)),
    np.nanmax(np.abs(uasmax_norm_hov.values)),
    np.nanmax(np.abs(diff_norm_hov.values)),
]))

fig, axes = plt.subplots(2, 3, figsize=(14, 10), constrained_layout=True)

im = plot_hov(axes[0, 0], uas_hov, "uas Hovm철ller (raw)", vlim=raw_vlim)
plt.colorbar(im, ax=axes[0, 0], label="m/s")

im = plot_hov(axes[0, 1], uasmax_hov, "uasmax Hovm철ller (raw)", vlim=raw_vlim)
plt.colorbar(im, ax=axes[0, 1], label="m/s")

im = plot_hov(axes[0, 2], diff_hov, "uasmax - uas (raw)", vlim=raw_vlim)
plt.colorbar(im, ax=axes[0, 2], label="m/s")

im = plot_hov(axes[1, 0], uas_norm_hov, "uas Hovm철ller (own-z)", vlim=norm_vlim)
plt.colorbar(im, ax=axes[1, 0], label="z-score")

im = plot_hov(axes[1, 1], uasmax_norm_hov, "uasmax Hovm철ller (own-z)", vlim=norm_vlim)
plt.colorbar(im, ax=axes[1, 1], label="z-score")

im = plot_hov(axes[1, 2], diff_norm_hov, "uasmax - uas (own-z)", vlim=norm_vlim)
plt.colorbar(im, ax=axes[1, 2], label="z-score difference")

plt.show()
```
